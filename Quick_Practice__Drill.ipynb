```python
import sys
# Two Pointer -- Remove duplicates from a sorted array
arr1 = [1,2,2,2,3,3,3,4,4,4,5,5,5,5,6,6,6,7,8,9,10,10,10,10]

i=0
for j in range(1,len(arr1)):
    if arr1[j] != arr1[i]:
        i+=1
        arr1[i]=arr1[j]
arr1 = arr1[:i+1]

# arr1= list(set(arr1))
print(arr1)
print('--------------------------------------------')

# Sliding Window -- Find the maximum sum of any 3 consecutive elements in an array.
arr2 = [2,4,1,5,6,3,-1,43]
k=3
n=len(arr2)
max_sum = -sys.maxsize

for i in range(0,n-k+1):
    curr_sum = 0
    for j in range(k):
        curr_sum += arr2[i+j]
    
    max_sum = max(curr_sum,max_sum)

print(arr2)
print('Max subarray sum of 3 consecutive element')
print(max_sum)
print('--------------------------------------------')


# Searching + 2 Pointer -- Given a sorted array and a target T, check if any 2 numbers sum to T (not necessarily return index).
arr3 = sorted([2,4,1,5,6,3,-1,43,2,4,2,6,5]) 
T = 11
def num_sum(arr, target):
    left, right = 0, len(arr) - 1
    while left < right:
        s = arr[left] + arr[right]
        if s == target:
            return True
        elif s < target:
            left += 1
        else:
            right -= 1
    return False

if num_sum(arr3, T):
    print('yes')
else:
    print('no')

```

```python
import re
# Check Palindrome
strr = "AyanayA"
if strr == strr[::-1]:
    print('Palindrom')


#  anagrams
str1 = 'Ayan Das'
str2 = 'Das Ayan'

str1_clean = str1.replace(' ','').lower()
str2_clean = str2.replace(' ','').lower()
if sorted(str1_clean) == sorted(str2_clean):
    print('The Strings are anagrams')
else:
    print('The Strings are not anagrams')

# Count Character
str1 = 'Ayan DasAyan'
dict_char = {}
for ch in str1:
    if ch not in dict_char:
        dict_char[ch] = 1
    elif ch in dict_char:
        dict_char[ch] += 1
print('Count Characterv :')
print(dict_char)

# Remove spaces and special characters
str_special = ' Ayan @& Das ^2*'
str_clean = re.sub(r'[^a-zA-Z0-9]', '',str_special)
print(str_clean)

s = "Ayan!%6&#@ 123."
res = ''
for char in s:
    if char.isalnum():
        res += char
print(res)
      

```
