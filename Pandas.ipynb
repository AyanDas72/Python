## Pandas

This is a library used for analysis and cleaning two Data Structure -- 1.Series & 2.Data Fram

```python
import pandas as pd
```

#### Basics

```python
##### Read CSV & Inspect #####
# Using the CSV file you downloaded (data.csv):

# Load it using pd.read_csv()
df = pd.read_csv('Data_&_text/data.csv')
print(df)

print("######################")
# Print the first 3 rows (head())
print(df.head(3))

print("######################")
# Print the last 3 rows 
print(df.tail(2))

print("######################")
#Get number of rows & columns
print(df.shape)

print("######################")
#Name of columns
print(df.columns)

print("######################")
#Datatype of columns
print(df.dtypes)

print("######################")
# Print summary statistics (describe())
df.describe()
```

```python
##### Cleaning & Filtering #####
# From the CSV:

#Check for missing values
import math


print(df.isnull().any(axis=1))
print(df.isnull().sum)

print("######################")
#Handling missing vallues
df_filled = df.fillna(0)
print(df_filled)

print("######################")
# Show all rows where City = “Kolkata”
print(df[df['City']=='Kolkata'])

print("######################")
# Show all rows where Age > 20
print(df[df['Age']>20])

print("######################")
# Show only Name + Age columns
df[['Name','Age']]

print("######################")
#Sorting by Value
df.sort_values('Age')

print("######################")
#Create a new column using simple arithmetic 
df_mean= df['Age'].mean() 
df['Base_Age_Difference'] = math.floor(df_mean)   - df['Age']
print(df)
```

```python
##### Aggregations #####
data1 = {
    "Name": ["Ayan", "Riya", "Sam", "Ayan", "Riya"],
    "Age": [25, 30, 35, 28, 32],
    "Score": [80, 90, 85, 70, 95]
}
df1 = pd.DataFrame(data1)

mean_age = df1["Age"].mean()
median_age = df1["Age"].median()
max_age = df1["Age"].max()
min_age = df1["Age"].min()

print("Mean:", mean_age)
print("Median:", median_age)
print("Max:", max_age)
print("Min:", min_age)

# Group by Name and compute aggregations
grouped = df1.groupby("Name").agg({
    "Age": ["mean", "count", "sum"],
    "Score": ["mean", "count", "sum"]
})

print(grouped)
```
