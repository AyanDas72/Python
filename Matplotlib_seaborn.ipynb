```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Data_&_text/BMW sales data (2010-2024) (1).csv')
#print(df)



# x=[1,2,3,4,5]
# y=[1,4,9,16,25]
# plt.plot(x,y)

yearly_sales = df.groupby("Year")["Sales_Volume"].mean().reset_index()
# print(yearly_sales)
# plot Sales_Volume vs Year
plt.plot(yearly_sales['Year'],yearly_sales['Sales_Volume'])
plt.xlabel("Avg Sales Volume")
plt.ylabel("year")
plt.title("Basic Line Plot")
plt.show()

#Histrogram
plt.hist(df["Sales_Volume"], bins= 10, edgecolor='black', color='Orange')
plt.title("Distribution of Sales Volume")
plt.xlabel("Sales Volume")
plt.ylabel("Frequency")
plt.show()

# bar plot
df.groupby('Model')['Sales_Volume'].mean().plot(kind='bar')
plt.title("Average Sales by Model")
plt.xlabel("Model")
plt.ylabel("Avg Sales Volume")
plt.show()
#type 2
avg_sales = df.groupby("Region")["Sales_Volume"].mean()
categories = avg_sales.index
values = avg_sales.values
plt.bar(categories, values)
plt.title("Average Sales Volume by Region")
plt.xlabel("Region")
plt.ylabel("Avg Sales Volume")
plt.xticks(rotation=45)
plt.show()

plt.scatter(categories, values, alpha=0.5)
plt.title("Region vs Avg Sales")
plt.xlabel("Region")
plt.ylabel("Avg Sales")
plt.show()

plt.boxplot(df['Price_USD'])
plt.title("Price Distribution (Box Plot)")
plt.ylabel("Price (USD)")
plt.show()

# Seaborn use 
sns.barplot(x='Fuel_Type', y='Sales_Volume', data=df)
plt.title("Sales by Fuel Type")
plt.show()

sns.countplot(x='Region', data=df)
plt.title("Count of Cars by Region")
plt.show()

sns.pairplot(df[['Engine_Size_L', 'Mileage_KM', 'Price_USD']])
plt.show()

corr = df[['Engine_Size_L', 'Mileage_KM', 'Price_USD', 'Sales_Volume']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
